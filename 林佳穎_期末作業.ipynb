各大學在第一至第三類學群只能錄取同一推薦高中學生一名；而第四至第七類學群則可分別錄取同一推薦高中學生各一名；第八類學群則是在同一所高中的學生中只有一名可通過篩選。

我了解了。那第二輪分發呢？

若第一輪分發後校系仍有缺額，會再針對第一輪分發未獲錄取的學生進行缺額校系分發作業，各大學在這次分發時就不受同一推薦學校錄取人數一名的限制了。

原來如此，所以如果被分發到就成為大學生了對吧！

第一到第七學群確實如此，不過第八類學群申請醫學系或牙醫學系的學生，基於醫師養成的特殊需求，按規定必須參加面試。

如果參加（繁星推薦）有什麼特別要注意的地方嗎？

（繁星推薦）的錄取生無論是否放棄，都不能參加當年度個人申請招生，這點一定要注意。

3.	個人申請

（個人申請）應該可以說是相較於過去的升學管道，改變最多的一個了。

天啊，太可怕了。

不過你也不用害怕，我會詳細地跟你說明。

108課綱的學測時間，依樣是在一月底舉行，不過呢，申請入學的招生作業時間延至高三課程結束，也就是差不多在五月辦理。

咦，為什麼要做這樣的調整？

其實就是因為希望讓學生在高三階段的學習能更完整。

除了這個以外，（個人申請）跟以前還有什麼不同嗎？

當然啊，以前的學測都是舉行兩天，自111學測開始將舉行三天。

那真是很考驗學生的毅力呢！在考試科目上有做什麼調整嗎？

考試科目從以前的五顆選考改為六科，其中將數學分成數學A、數學B兩種試卷。

命題方式又有什麼不同呢？

主要是持續推動素養導向命題，不過除了選擇題外，各考科皆會增加非選擇題的部分，以混合題型來評量學生的表達說明能力，同時將答案卡與試題卷合併為同一張答題卷。

聽說以前的備審資料在課綱更改後變成學習歷程檔案，那又是什麼意思啊？

學習歷程檔案主要分成四大部分，包括基本資料、修課紀錄、課程學習成果跟多元表現。

對於學習歷程檔案有什麼特殊的要求嗎？

多數是要在每學期上傳，在（個人申請）第二階段申請時再至中央資料庫選擇並勾選上傳，當然上傳時也會也有所條件限制。其他項目則是依據各大學校系規定，在報名時上傳。

你說的限制例如什麼呢？

像是課程學習成果每學年上傳至多六件，同時內容需要有老師認證，而多元表現每學年上傳至多十件。

關於學測成績跟學習歷程檔案的比例是怎麼分配的？

每個校系通常不太一樣，不過有統一規定學測佔比不能超過50%喔，而且書審及口試委員是看不到學生學測成績的。

4.分科測驗

（分科測驗）相較以往有蠻大幅度的變動，讓我來跟你說明。

好的，麻煩你了。

測驗的時間跟以往一樣維持在七月初，但考試科目則由十科減為七科。

咦？是哪七科呢？

包括數學甲、物理、化學、生物、歷史、地理還有公民。

國文、英文、數學乙怎麼不見了？

這三科的分數就以學測成績為主喔，沒有第二次考試機會了。

天啊，那根本是一次定生死。

沒錯，至於測驗題型呢，包括選擇題、選填題型跟非選擇題型，跟學測一樣是混合題型喔。

那計分方式跟之前一樣嗎？

完全不一樣，（分科測驗）也採用級分制而不是百分制了。

我了解了，真的謝謝你喔。

不會，這次的變動可以說是近年來最大的一次，面對新的招考制度，雖然沒有任何經驗可以參考獲續會讓你感到迷茫，不過如果你做好功課，了解深入108課綱，多少能夠減少一些因為面對新制度所產生的焦慮以及不安。
